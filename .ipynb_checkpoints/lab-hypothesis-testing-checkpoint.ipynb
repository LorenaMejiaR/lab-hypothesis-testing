**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import ttest_ind
from scipy.spatial.distance import euclidean
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

```python
df.columns
```

```python
df.columns= ['Name', 'Type_1', 'Type_2', 'HP', 'Attack', 'Defense', 'Sp_Atk',
       'Sp_Def', 'Speed', 'Generation', 'Legendary']
df.columns
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#H0: Pokemons Dragon HP <= Pokemons Grass HP
#H1: Pokemons Dragon HP > Pokemons Grass HP

#significance level = 0.05
```

```python
#code here
dragon_hp = df[df['Type_1'] == 'Dragon']['HP']
grass_hp = df[df['Type_1'] == 'Grass']['HP']
```

```python
st.ttest_ind(dragon_hp,grass_hp, equal_var=False, alternative="greater")
```

### Because p_value (0.0007993609745420597) is lower than significance level (0.05) , we reject the null hypothesis. 
### This means that the Dragon have more HP than Grass. 

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# Create the LEGENDARY and NOT-LEGENDARY columns
df['legendary'] = df['Legendary'].apply(lambda x: True if x else False)
df['non_legendary'] = df['Legendary'].apply(lambda x: False if x else True)
df
```

```python
#code here
alpha = 0.05
num_tests = len(['HP', 'Attack', 'Defense', 'Sp_Atk', 'Sp_Def', 'Speed'])
alpha_bonferroni = alpha / num_tests 

for stat in ['HP', 'Attack', 'Defense', 'Sp_Atk', 'Sp_Def', 'Speed']:
    stat_leg = df[df['Legendary'] == True][stat]
    stat_nonleg = df[df['Legendary'] == False][stat]
    
    t_stat, p_value = ttest_ind(stat_leg, stat_nonleg)
    
    if p_value < alpha_bonferroni:
        print(f"{stat}: Significant difference (p-value = {p_value})")
    else:
        print(f"{stat}: No significant difference (p-value = {p_value})")
```

### Since all the p-values are extremely small, we can confidently reject the null hypothesis for all six stats. 
## This means that there are significant differences between Legendary and Non-Legendary Pokémon.

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
school_coords = (-118, 37)
hospital_coords = (-122, 34)

# Function to calculate Euclidean distance
def calculate_distance(x1, y1, x2, y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
```

```python
# Add columns for distance to school and hospital
def add_distance_columns(df2):
    df2['distance_to_school'] = df2.apply(lambda row: calculate_distance(row['longitude'], row['latitude'], *school_coords), axis=1)
    df2['distance_to_hospital'] = df2.apply(lambda row: calculate_distance(row['longitude'], row['latitude'], *hospital_coords), axis=1)
    return df2
```

```python
# Classify houses into close and far
def classify_houses(df2, threshold=0.50):
    df2['close_to_school_or_hospital'] = df2.apply(
        lambda row: 'close' if (row['distance_to_school'] < threshold or row['distance_to_hospital'] < threshold) else 'far', axis=1)
    return df2
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df2.columns)
```

```python
print(df2.head())
```

```python
# Add the distance columns
df2 = add_distance_columns(df2)

# Classify the houses
df2 = classify_houses(df2)
df2.describe()
```

```python

# Now, perform a t-test to compare the median_house_value of the two groups
close_group = df2[df2['close_to_school_or_hospital'] == 'close']['median_house_value']
far_group = df2[df2['close_to_school_or_hospital'] == 'far']['median_house_value']

# Perform a t-test
t_stat, p_value = st.ttest_ind(close_group, far_group, equal_var=False)

# Output the result
if p_value < 0.05:
    result = f"Reject the null hypothesis (p-value = {p_value:.4f}). Houses close to a school or hospital have significantly higher median house values."
else:
    result = f"Fail to reject the null hypothesis (p-value = {p_value:.4f}). There is no significant difference in house values."

print(result)
```
